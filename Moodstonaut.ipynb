```python
%%html
<style>
  /* starry background */
  .starry {
    position:fixed; top:0; left:0; width:100%; height:100%; pointer-events:none; z-index:-1;
  }
  .star {
    position:absolute; width:2px; height:2px; background:#FFF;
    animation:twinkle 3s infinite ease-in-out;
  }
  @keyframes twinkle {0%,100%{opacity:0.6;}50%{opacity:1;}}

  /* container */
  #mood-app {
    font-family:'Moon Sans',sans-serif; color:#F4EFFF;
    display:flex; align-items:center; justify-content:center;
    height:100vh; margin:0; padding:0; background:#14162C;
  }
  .container {
    position:relative; z-index:2; width:360px; text-align:center;
    background:rgba(29,32,51,0.85); padding:30px; border-radius:16px;
    box-shadow:0 8px 24px rgba(0,0,0,0.5);
  }
  .logo { font-size:2.4rem; margin-bottom:20px; }
  .tabs { display:flex; justify-content:space-around; margin-bottom:20px; }
  .tab {
    cursor:pointer; padding:8px 16px; color:#F4EFFF;
  }
  .tab.active { border-bottom:2px solid #F4EFFF; }
  .form-group { margin-bottom:16px; text-align:left; }
  .form-group label { display:block; color:#D1D6E8; margin-bottom:6px; }
  .form-group input[type="email"],
  .form-group input[type="password"],
  .form-group textarea {
    width:100%; padding:10px; border:none; border-radius:12px;
    background:rgba(42,45,68,0.8); color:#F4EFFF;
  }
  .btn {
    width:100%; padding:12px; background:#4A4D8C; border:none;
    border-radius:12px; color:#F4EFFF; font-weight:bold; cursor:pointer;
    margin-top:8px;
  }
  .link { display:block; margin-top:8px; color:#B5A7D9; text-decoration:none; }
  .link:hover { text-decoration:underline; }
  #msg { min-height:1.2em; margin-bottom:12px; color:#FFBABA; }
  /* Survey */
  .survey qgroup { margin-bottom:16px; display:block; text-align:left; }
  .survey .options { display:flex; justify-content:space-around; }
  .survey .options label { cursor:pointer; }
  .survey input[type="range"] { width:100%; }
</style>

<div id="mood-app">
  <div class="starry">
    <div class="star" style="top:10%; left:20%;"></div>
    <div class="star" style="top:30%; left:80%; animation-delay:1s;"></div>
    <div class="star" style="top:70%; left:15%; animation-delay:2s;"></div>
    <div class="star" style="top:50%; left:60%;"></div>
  </div>

  <div class="container">
    <div class="logo">Moodstronaut</div>
    <div class="tabs">
      <div id="tab-login" class="tab active">Login</div>
      <div id="tab-register" class="tab">Register</div>
      <div id="tab-survey" class="tab">Survey</div>
    </div>

    <div id="msg"></div>

    <!-- LOGIN FORM -->
    <form id="login-form">
      <div class="form-group">
        <label for="login-email">Email</label>
        <input id="login-email" type="email" placeholder="example@domain.com">
      </div>
      <div class="form-group">
        <label for="login-password">Password</label>
        <input id="login-password" type="password" placeholder="••••••••">
      </div>
      <button type="button" class="btn" id="btn-login">Log in</button>
      <a href="#" class="link" id="link-to-register">Register</a>
    </form>

    <!-- REGISTER FORM -->
    <form id="register-form" style="display:none;">
      <div class="form-group">
        <label for="reg-email">Email</label>
        <input id="reg-email" type="email" placeholder="example@domain.com">
      </div>
      <div class="form-group">
        <label for="reg-password">Password</label>
        <input id="reg-password" type="password" placeholder="••••••••">
      </div>
      <div class="form-group">
        <label for="reg-confirm">Confirm Password</label>
        <input id="reg-confirm" type="password" placeholder="••••••••">
      </div>
      <button type="button" class="btn" id="btn-register">Register</button>
      <a href="#" class="link" id="link-to-login">Back to Login</a>
    </form>

    <!-- SURVEY FORM -->
    <form id="survey-form" class="survey" style="display:none;">
      <!-- Q1: Sky type -->
      <div class="form-group">
        <label>What kind of sky was your day under?</label>
        <div class="options">
          <label><input type="radio" name="sky" value="sunny"> ☀️ Sunny</label>
          <label><input type="radio" name="sky" value="cloudy"> ☁️ Cloudy</label>
          <label><input type="radio" name="sky" value="stormy"> ⚡️ Stormy</label>
        </div>
      </div>
      <!-- Q2: Emotion -->
      <div class="form-group">
        <label>What emotion pulled you today?</label>
        <div class="options">
          <label><input type="radio" name="emotion" value="joy"> Joy</label>
          <label><input type="radio" name="emotion" value="anxiety"> Anxiety</label>
          <label><input type="radio" name="emotion" value="curiosity"> Curiosity</label>
          <label><input type="radio" name="emotion" value="grief"> Grief</label>
        </div>
      </div>
      <!-- Q3: Grounded → Lost slider -->
      <div class="form-group">
        <label>Where were you emotionally?</label>
        <input type="range" id="slider-ground" min="0" max="100" value="50">
        <div style="display:flex; justify-content:space-between; color:#D1D6E8;">
          <small>Grounded</small><small>Lost</small>
        </div>
      </div>
      <!-- Q4: Whisper -->
      <div class="form-group">
        <label>Would you like to leave a whisper in stars?</label>
        <textarea id="whisper" rows="3" placeholder="Your message…"></textarea>
      </div>
      <button type="button" class="btn" id="btn-survey">Submit Mood</button>
    </form>
  </div>
</div>

<script>
(function(){
  // Simple in-memory store for login/register
  let users = JSON.parse(localStorage.getItem('mood_users')||'{}');

  // Element refs
  const tabs      = {
          login:    document.getElementById('tab-login'),
          register: document.getElementById('tab-register'),
          survey:   document.getElementById('tab-survey')
        },
        forms     = {
          login:    document.getElementById('login-form'),
          register: document.getElementById('register-form'),
          survey:   document.getElementById('survey-form')
        },
        msg        = document.getElementById('msg'),
        buttons    = {
          login:    document.getElementById('btn-login'),
          register: document.getElementById('btn-register'),
          survey:   document.getElementById('btn-survey')
        },
        links      = {
          toReg:  document.getElementById('link-to-register'),
          toLogin:document.getElementById('link-to-login')
        },
        dashboard  = document.getElementById('dashboard-app'),
        moodApp    = document.getElementById('mood-app'),
        userNameEl = document.getElementById('user-name');

  // Utility to show one form & activate its tab
  function showForm(which){
    for(let k in forms){
      forms[k].style.display = (k===which)? 'block':'none';
      tabs[k].classList.toggle('active', k===which);
    }
    msg.textContent = '';
  }

  // LOGIN handler
  function doLogin(){
    const e = document.getElementById('login-email').value.trim(),
          p = document.getElementById('login-password').value;
    if(users[e]===p){
      // On success, show dashboard
      userNameEl.textContent = e;
      moodApp.style.display      = 'none';
      dashboard.style.display    = 'flex';
    } else {
      msg.style.color = '#FFBABA';
      msg.textContent = 'Invalid email or password.';
    }
  }

  // REGISTER handler
  function doRegister(){
    const e=document.getElementById('reg-email').value.trim(),
          p=document.getElementById('reg-password').value,
          c=document.getElementById('reg-confirm').value;
    if(!e||!p){
      msg.style.color='#FFBABA'; msg.textContent='Please fill all fields.'; return;
    }
    if(p!==c){
      msg.style.color='#FFBABA'; msg.textContent='Passwords do not match.'; return;
    }
    if(users[e]){
      msg.style.color='#FFBABA'; msg.textContent='Email already exists.'; return;
    }
    users[e]=p;
    localStorage.setItem('mood_users', JSON.stringify(users));
    // Persist via Python if available
    if(window.IPython && IPython.notebook && IPython.notebook.kernel){
      const code = `save_user(${JSON.stringify(e)}, ${JSON.stringify(p)})`;
      IPython.notebook.kernel.execute(code);
    }
    msg.style.color='#B5FFA7'; msg.textContent='Registration successful!';
    setTimeout(()=>showForm('login'), 800);
  }

  // SURVEY handler
  function doSurvey(){
    const sky     = document.querySelector('input[name="sky"]:checked'),
          emo     = document.querySelector('input[name="emotion"]:checked'),
          ground  = document.getElementById('slider-ground').value,
          whisper = document.getElementById('whisper').value.trim();
    if(!sky||!emo){
      msg.style.color='#FFBABA'; msg.textContent='Please answer at least the first two questions.'; return;
    }
    // For now we simply log the answers
    console.log({
      sky: sky.value,
      emotion: emo.value,
      grounding: ground,
      whisper
    });
    msg.style.color='#B5FFA7';
    msg.textContent='Thanks! Your mood has been recorded.';
    // clear
    document.getElementById('survey-form').reset();
    setTimeout(()=>showForm('survey'), 800);
  }

  // Event bindings
  tabs.login.addEventListener('click', ()=>showForm('login'));
  tabs.register.addEventListener('click', ()=>showForm('register'));
  tabs.survey.addEventListener('click', ()=>showForm('survey'));
  buttons.login.addEventListener('click', doLogin);
  buttons.register.addEventListener('click', doRegister);
  buttons.survey.addEventListener('click', doSurvey);
  links.toReg.addEventListener('click', e=>{ e.preventDefault(); showForm('register'); });
  links.toLogin.addEventListener('click', e=>{ e.preventDefault(); showForm('login'); });

  // init
  showForm('login');
})();
</script>
```
